```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df  = pd.read_csv('Customer_Churn.csv')
df
```

# converted 0 and 2 values of seniorcitizen to yes/no to make it easier to underrstand 

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example data (use your own df)
# df = your dataframe

# Step 1: Create a cross-tab (frequency table)
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize='index') * 100

# Step 2: Plot stacked bar chart
ct.plot(kind='bar', stacked=True, figsize=(4,4))

# Step 3: Add percentage labels
for i, row in enumerate(ct.values):
    cumulative = 0
    for value in row:
        plt.text(i, cumulative + value/2, f"{value:.1f}%", ha='center', va='center', color='white', fontsize=10)
        cumulative += value

# Step 4: Style
plt.title("Churn by SeniorCitizen (%)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.legend(title="Churn" , bbox_to_anchor = (0.9,0.9))
plt.show()
```

```python
# Service-related categorical columns
service_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the figure size and layout (3 rows x 3 columns)
plt.figure(figsize=(16, 12))

# Loop through columns and create subplots
for i, col in enumerate(service_cols, 1):
    plt.subplot(3, 3, i)  # 3x3 grid
    ax = sns.countplot(data=df, x=col , hue = "Churn")
    
    # Add percentage labels
    total = len(df)
    for container in ax.containers:
        labels = [f'{(v.get_height() / total * 100):.1f}%' if v.get_height() > 0 else '' for v in container]
        ax.bar_label(container, labels=labels, label_type='edge', fontsize=8)
    
    plt.title(f"count plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=20)
    plt.tight_layout()

plt.show()
```

```python
# he visualizations show that customers with additional services like OnlineSecurity, TechSupport, and DeviceProtection are less likely to churn, while those without these services have a higher churn rate. 📊
# Churn is also significantly higher among customers using month-to-month contracts and those with Fiber optic internet.
# Overall, offering bundled or long-term service plans seems to improve customer retention.
```
